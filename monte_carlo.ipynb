## Background

Monte Carlo methods are essentially random (or stochastic) methods. We take random samples of our problem and the central idea is that the more random samples we take, the closer our random method will come to the true solution.

Monte Carlo methods are very diverse and have applications from solving differential equations to optimization method to even quantum chromodynamic calculations. More classically these methods are used in numerical integration, and are especially powerful in multidimensional integration as quadrupling the number of random samples will halve the error, whatever of the number of dimensions.

… In case you were wondering, the name ‘Monte Carlo’ was given to the method in honour of one of the pioneer’s uncles, who was a gambler. The name itself refers to the famous casinos in the Monaco quarter. Before this, the methods were referred to by slightly less glamorous names such as ‘statistical sampling’.

## Approximating $\pi$

To introduce the Monte Carlo method we’re first going to use it calculate the value of $\pi$. Imagine that you're a really bad darts player. So bad, in fact, that each dart lands in a random place within a square that encloses the dart board, (figure 1).

Now the probability of the dart landing within the circular dartboard, $P_c$, is the area of the dartboard $A_c$ divided by the area of the enclosing square, $A_s$.

$P_c = \frac{A_c}{A_s}$

Let us call the radius of the dartboard $r$, so the enclosing square has side $2r$. Then our probability $P_c$ is:

$P_c = \frac{\pi r^2}{4 r^2} = \frac{\pi}{4}$ (1)

Now, if we throw N darts, the probability of the dart landing within the dartboard is equal to the number of darts that end up inside the dartboard Nc, divided by the number of darts landing within the square. However, the number of darts within the square equals the total number of darts, so:

$P_c \approx \frac{N_c}{N} \approx \frac{\pi}{4}$

Where the last equality comes from (1). From this we can see that an estimate for $\pi$ is:

$\pi \approx 4 \cdot \frac{N_c}{N}$ (2)

As $N \rightarrow \infty$, our estimate for $\pi$ will approach the true value of $\pi$.

### Plotting a quarter of a circle

Next, we define x and y coordinates that correspond to the edge of one quarter of a circle, that means, all points that satisfy $x^2 + y^2 = r^2$, in our case $r$ being equal to $1$, we get $y^2 = 1 - x^2$. Thus, for a given $x$ value we can find the corresponding y value using $y = \sqrt{1 - x^2}$. Therefore, we calculate our $y$ coordinates for 100 evenly spaced $x$ coordinates between 0 and 1. NumPy has a convenient feature called vectorization – that means, when you apply a function that is normally applied to a scalar (such as square, subtraction and square root, in this case), it is applied elementwise. It is also noteworthy that `1 - circle_x ** 2` is a subtraction of an array from a scalar, the feature that allows this (and much more) is called broadcasting [1] in NumPy, but for the purposes of this exercise it is enough to know that it works.

```python
circle_x = np.linspace(0, 1, 100)
circle_y = np.sqrt(1 - circle_x ** 2)
```

Next, we need to set up a squared figure (in this case, 5 by 5 inches – though this is not all that meaningful related to display output, it is just important that it is square) which we then use to plot the circle on.

```python
fig = plt.figure()
fig.set_size_inches(5, 5)
plt.plot(circle_x, circle_y)
```

### Plotting random points

To simplify maintainment, we define the amount of random points we place on the grid in a variable, in this case 2000, which will only roughly average $\pi$, but which prevents the following plots becoming too cluttered.

```python
TIMES = 2000
```

Then, we use the `np.random.rand(n)` method, which produces an array containing n uniformly distributed random numbers in [0, 1), to generate x and y coordinates for our random points.

```python
x = np.random.rand(TIMES)
y = np.random.rand(TIMES)
```

We then plot the random points as points (using the 'o' argument to plot prevents it from plotting a line – `markersize=1 makes the points smaller) and then overplot the circle again.

```python
fig = plt.figure()
fig.set_size_inches(5, 5)
plt.plot(x, y, 'o', markersize=1)
plt.plot(circle_x, circle_y)
```

### Calculating the distances

We then, for each of our
points, check whether it is within the circle – that is, whether its Euclidean distance from the origin is smaller or equal 1. So, we apply $\sqrt{x^2 + y^2} \le 1$ to each of our points, which gives us back an array containing truth values for each of our points. This implementation is not optimal though – see if you can come up with a better one.

```python
dist = np.sqrt(x ** 2 + y ** 2)
```

```python
incircle = dist <= 1
```

### Calculating $\pi$

The percentage of random points that are within the circle can be obtained by counting the amount of points where $\sqrt{x^2 + y^2} \le 1$ evaluates true and dividing by the total number of random points. We need to convert either of the two numbers to a floating point number because Python division of two integers rounds down to the next integer – in our case almost certainly 0.

```python
incircle_ratio = incircle.sum() / float(incircle.size)
```

We can now calculate $\pi$, as discussed above, by multiplying that percentage by 4.

```python
pi = incircle_ratio * 4
```

```python
print pi
```

### Visualize the error

Now, let’s visualize the approximation by keeping track of the error in our approximation of $\pi$ as we increase the number of samples. In the Monte Carlo method the error is roughly proportional to the standard deviation of our set of approximate values. If you haven’t met standard deviation before, it’s simply a measure of how widely spread a distribution of values is around the mean of the same set of values. For that, we first need to approximate $\pi$ for the different sample sizes. We do that using `np.cumsum`, which stands for cumulative sum, that means, the element at index i is the sum of all elements up to that index, in our case the number of “darts” that hit our circle having thrown `i + 1` (Python is zero-based indexed) of them.

```python
cumulative_incircle = np.cumsum(incircle)
```

Thus, to get the ratio of hit/thrown after `n` throws, all we need to do is divide each element of the cumulative sum by its `index + 1`. We do that by generating an array that goes from 1 to `TIMES + 1` excluded using `np.arange`; its data-type must be set to float so the result of the division turns out float as well.

```python
cumulative_ratios = (
    cumulative_incircle / np.arange(1, TIMES + 1, dtype=np.float)
)
```

We can then generate the value approximated for $\pi$ after `n` throws by multiplying the ratios by 4, as discussed earlier.

```python
pis = cumulative_ratios * 4
```

We can then use the generated data to plot the approximated value of $\pi$ versus sample size. We're also adding a reference line at the actual value of $\pi$, which we have to repeat `TIMES` times so that it spans the whole figure.

```python
plt.figure()
approx_pis, = plt.plot(pis)
pi, = plt.plot(np.repeat(np.pi, TIMES))
# Restrict y-axis to [3.1, 3.2] so that more detail is visible.
plt.ylim(3.1, 3.2)
plt.xlabel("Sample size")
plt.legend([approx_pis, pi], ["Approximation", "Exact value"])
```

We then calculate and plot the the standard deviation of approximized values for $\pi$ for every sample size, that means, the value at index `n` is the standard deviation of all elements up to index `n` in the `pis` array. To achieve that, we program two small functions. The standard deviation is, per definition, the square-root of the variance which is the mean of the squares minus the square of the means; we can use that definition to first implement a function `cummean` that calculates the mean of all values up to `n` in the `n`-th element, and then use that to implement `cumstd`. Functions are defined by using the `def` keyword, followed by a space and the name of the function, its arguments in parentheses, a colon and the method body intended by four spaces.

```python
def cummean(arr):
    return np.cumsum(arr) / np.arange(1, len(arr) + 1, dtype=np.float)

def cumstd(arr):
    return np.sqrt(cummean(arr ** 2) - cummean(arr) ** 2)

plt.figure()
stdevs, = plt.plot(cumstd(pis))
plt.legend([stdevs], ["Standard deviation"])
plt.xlabel("Sample size")
```

## Numerical integration

Being able to calculate the value of $\pi$ is one thing, but we can use the same method to do something much more useful (to physics at least); we can use it to numerically calculate the value of definite integrals.

The way we can do this is almost exactly the same as above, apart from instead of a round dartboard we are throwing our ‘darts’ at a function and determining whether the random ‘hit’ lies above or below the function. By taking the ratio of the areas and equating it with the probability of the dart landing in each area we can solve the integral numerically.

Our relation still holds:

$P_c = \frac{N_f}{N} \approx \frac{A_f}{A}$

Where N is the total number of samples, $N_f$ is the number of samples under the function, $A_f$ is the area under the function and A is the total bounding area. Since the area under the function is simply the integral, this gives:

$\int_a^b f(x) \mathrm{d}x  \approx (b - a) \cdot [f(b) - f(a)] \cdot \frac{N_f}{N}$

The steps defined above can be used almost unchanged to approximate the definite integral of a function, in our case $f(x) = x^2$.
