## Background

Electrostatic forces can be expressed using the concept of an electric fields. The electric
field, $\mathbf E$, is a vector field in which the vectors at each point in space are given by the
force, $\mathbf F$, that would be present on a positive test charge at that point:

$\mathbf F = q\mathbf E$

Where q is the charge of the small test charge. The electric field of a point charge Q is
given by:


$\mathbf E = k\frac{Q}{\|\textbf r\|^2} \hat{\textbf r}$

Where $\|\textbf r\|$ is the distance from the charge $Q$ to the point at which the field is measured and
$\hat{\mathbf r}$ is a unit vector pointing from Q to the evaluation point. Now, substituting (2) into (1)
gives the force of attraction between two point charges as:


$\mathbf F = k\frac{qQ}{\|\textbf r\|^2} \hat{\textbf r}$

This is Coulomb's law. $\mathbf F$ is the electrostatic force on charge $q$ caused by charge $Q$, $k$ is
the electrostatic constant, and $\hat{\mathbf r}$
 is a unit vector pointing from $Q$ to $q$.


The electric field is a vector field, but can be simplified even further by using electric
potential. The electric potential is a scalar field, so it only needs one value for each point
in space, unlike the electric field which requires three values (the x, y and z values of a
vector). The potential is given by:


$\varphi = k \frac{Q}{\|\mathbf r\|}$

And the electric field is then given by:


$E = -\nabla \varphi$

Where the upside down triangle, ∇, is the gradient operator:

$\nabla = \frac{\delta}{\delta x} \hat{\mathbf i} + \frac{\delta}{\delta y} \hat{\mathbf j}$

Just like the physical field, $\mathbf E$, the potential follows the *principle of superposition*; so if
we have the potential of one unit charge, we can find the potential of many unit charges,
by simply adding scalar values. Once we have the potential, we can then find the electric
field of the charge distribution by simply taking the gradient; there's no need to add up all
the individual electric fields. Extending the idea, we can find the electric field produced
by a continuous amount of charge, by integrating a **charge density** instead of summing
**discrete charges**.


We can draw an analogy with potentials and fields by trying to imagine the potential as a
hill. A ball placed on a hill will roll in the downhill direction; this direction is the
gradient of the hill. The figure above shows the hill (as the circles, imagine it as a
contour map with the highest point in the centre) and the directions a ball would take if
placed on various the hill (as the vectors). Now, we only need one scalar value to
describe the hill at any point (the height) but by taking the gradient we can find the
direction the ball would travel in. This is analogous to the electrostatic situation, the
direction of the electric force (a vector) can be represented by the gradient of a scalar (the
potential).


## Discrete point charges

For this exercise, additionally to `NumPy` and `matplotlib`, we will also need `sympy`, a package for symbolic math in Python, and `scipy`, a package for advanced numeric math (builing on `NumPy`).

```python
import numpy as np
import scipy.integrate
from matplotlib import pyplot as plt

import sympy
```

We start by defining a function for the potential of a single charge at the origin. We've set $k=1$ and $q=1$ here, this is alright as long as we remember that the vector field and potential are scaled, and that we've set our charge to 1.

The syntax of function definitions has already been discussed in the previous exercise.

```python
def potential(x, y):
    return 1.0 / np.sqrt(x ** 2 + y ** 2)
```

We can then evaluate the function at arbitrary points.

```python
potential(0.4, -0.6)
```

```python
xs = ys = np.linspace(-2, 2, 200)
```

We then use the `np.meshgrid` function to generate two arrays for x and y coordinates that contain all points on the grid spanned by the two coordinate arrays.

```python
xv, yv = np.meshgrid(xs, ys)
```

```python
plt.imshow(
    potential(xv, yv),
    extent=[xs.min(), xs.max(), ys.min(), ys.max()]
)
plt.colorbar()
```

```python
fig = plt.figure()
fig.set_size_inches(5, 5)
cont = plt.contour(xs, ys, potential(xv, yv), [0.4, 0.5, 0.6, 1, 2, 3, 4, 5])
# Plot labels with one decimal point.
plt.clabel(cont, inline=1, fontsize=9, fmt='%0.1f')
```

Now comes the most complicated part of the exercise – visualizing the electric field as a vector field. We do this by evaluating eq. 5 (that is, taking the gradient of our potential function) at the points in our grid and then plotting them using the `plt.quiver` function. We use sympy to symbolically differentiate our potential function with respect to x and with respect to y to get two functions that give us the x and the y coordinate of our gradient vectors (`sympy.lambdify` turns the sympy expression into a Python function that receives x and y as parameters).

```python
vxs = np.linspace(-2, 2, 30)
vxv, vyv = np.meshgrid(vxs, vxs)

fig = plt.figure()
fig.set_size_inches(5, 5)

# potential as sympy expression
x, y = sympy.Symbol("x"), sympy.Symbol("y")
f = 1.0 / sympy.sqrt(x ** 2 + y ** 2)

# Calculate gradient
dx = sympy.lambdify([x, y], sympy.diff(f, x))
dy = sympy.lambdify([x, y], sympy.diff(f, y))

# Calculate x and y coordinates of vectors on our grid
vxs = -dx(vxv, vyv)
vys = -dy(vxv, vyv)

# Norm vectors
n = np.sqrt(vxs ** 2 + vys ** 2)
vxs = vxs / n
vys = vys / n

# Plot vectors
q = plt.quiver(vxv, vyv, vxs, vys, scale=35)
```

## Continuous Charge Distribution

If we need to work out the potential of a charged plate or wire, we could approximate the potential by breaking the wire into a number of charged sections of length x and summing the charge as shown above. As each of the charged sections becomes smaller and smaller, the approximation to the potential becomes more and more exact. If we take the limiting case of the length of each of the component sections becoming dx our summation will become an integration.

SymPy offers support for symbolic integration, while SciPy is required for numeric integration. Roughly speaking, an analytical solution is expressed in terms of 'simple' functions such as log, sine, cosine, etc. Numerical solutions use numerical methods to solve a definite integral and come up with an numerical answer for the integral.

We can, for instance, integrate the function $\frac{1}{4 + y^2}$.

```python
y = sympy.Symbol("y")
sympy.integrate(1 / sympy.sqrt(4 + y**2), y)
```

We can put limits into our integration by adding them to our variable.

```python
y = sympy.Symbol("y")
sympy.integrate(1 / sympy.sqrt(4 + y**2), (y, -4, 4))
```

Complicated integrals, like $\int \frac{1}{\sqrt{y \cos y}}$, cannot be solved analytically and we thus have to use the `scipy.integrate.quad` function to get a result.

```python
y = sympy.Symbol("y")
f = 1 / sympy.sqrt(y * sympy.cos(y))
# sympy.integrate(f, (y, 1, 1.5))
```

```python
s, e = -0.5, .5
px = sympy.Symbol("px")
py= sympy.Symbol("py")
f = (px - s) / sympy.sqrt((px - x) ** 2 + (py - y) ** 2)
f1 = f.subs(py, 0)

@np.vectorize
def potential(xv, yv):
    return scipy.integrate.quad(
        sympy.lambdify([px], f1.subs([(x, xv), (y, yv)]))
        , s, e
    )[0]

img = potential(xv, yv)
plt.imshow(img, extent=[xs.min(), xs.max(), ys.min(), ys.max()])
```

```python
fig = plt.figure()
fig.set_size_inches(5, 5)
cont = plt.contour(xs, ys, img, [0.4, 0.5, 0.6, 1, 2, 4])
# Plot labels with one decimal point.
plt.clabel(cont, inline=1, fontsize=9, fmt='%0.1f')
```

```python
df1x = f1.diff(x)
df1y = f1.diff(y)

@np.vectorize
def dx(xv, yv):
    return scipy.integrate.quad(
        sympy.lambdify([px], df1x.subs([(x, xv), (y, yv)]))
        , s, e
    )[0]

@np.vectorize
def dy(xv, yv):
    return scipy.integrate.quad(
        sympy.lambdify([px], df1y.subs([(x, xv), (y, yv)]))
        , s, e
    )[0]

fig = plt.figure()
fig.set_size_inches(5, 5)

# Calculate x and y coordinates of vectors on our grid
vxs = -dx(vxv, vyv)
vys = -dy(vxv, vyv)

# Norm vectors
n = np.sqrt(vxs ** 2 + vys ** 2)
vxs = vxs / n
vys = vys / n

# Plot every step-th vector
q = plt.quiver(vxv, vyv, vxs, vys, scale=35)
```
